```python
import matplotlib.pyplot as plt

models = ['SVM', 'DT', 'FT-BERT', 'Glove', 'GPT2', 'BLOOM', 'LAMMA2', 'GPT3.5', 'GPT4']
precision = [0.59, 0.57, 0.62, 0.58, 0.15, 0.37, 0.54, 0.67, 0.81]
recall = [0.59, 0.57, 0.62, 0.58, 0.39, 0.60, 0.53, 0.86, 0.57]
f1_score = [0.59, 0.57, 0.62, 0.58, 0.22, 0.46, 0.54, 0.75, 0.67]

# Plotting precision, recall, and F1-score for each model
plt.figure(figsize=(6, 4))

plt.plot(models, precision, marker='o', label='Precision')
plt.plot(models, recall, marker='s', label='Recall')
plt.plot(models, f1_score, marker='^', label='F1-Score')

plt.title('Model Evaluation Metrics')
plt.xlabel('Models')
plt.ylabel('Scores')
plt.ylim(0, 1)  # Set y-axis limit from 0 to 1 for score values
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

**BERT Model's Accuracy Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Model's accuracy on three different datasets

datasets = ['True/FAKE', 'WELFake', 'LiarDataset']
model_accuracy = [0.83, 0.83, 0.62]

# Set the width of the bars
bar_width = 0.25

# Set the x locations for the groups
index = np.arange(len(datasets))

# Plotting the grouped bar chart
plt.figure(figsize=(6, 4))

plt.bar(index, model_accuracy, bar_width, color='blue')

plt.xlabel('Datasets')
plt.ylabel('Accuracy')
plt.title('Model Performance on Different Datasets')
plt.xticks(index, datasets)
plt.ylim(0, 1)

# Show plot
plt.tight_layout()
plt.show()
```
